```python
import pickle
import torch
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load data from pickle file
pkl_file_kmeans = 'features_dict_kmeans_aug.pkl'
features_dict = pickle.load(open(pkl_file_kmeans, 'rb'))
print(f'Loaded {pkl_file_kmeans} ... ')

all_kmeans_cluster_center = {}

# Iterate over each label in the features dictionary
for label in features_dict:
    print(label)
    features = torch.cat(features_dict[label], dim=0)

    # Perform KMeans clustering
    kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(features)
    print(features)
    # Get cluster centers
    kmeans_cluster_center = torch.from_numpy(kmeans.cluster_centers_)
    all_kmeans_cluster_center[label] = kmeans_cluster_center
    print(all_kmeans_cluster_center[label])    

    # Visualization
    print(features[:, 0], features[:, 1])
    plt.scatter(features[:, 0], features[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.5)
    plt.scatter(kmeans_cluster_center[:, 0], kmeans_cluster_center[:, 1], c='red', marker='x', s=100)
    plt.title(f'Data and Cluster Centers for {label}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()
```
